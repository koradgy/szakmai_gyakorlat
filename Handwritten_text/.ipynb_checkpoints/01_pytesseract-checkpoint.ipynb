```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
import os
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
import pandas as pd
from PIL import Image
```

# Kézírásszerű betűtípus

```python

image_path = 'Data/Handwriten_font/page1.jpg'
image = cv2.imread(image_path)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Élkeresés a Canny algoritmussal
edges = cv2.Canny(gray_image, threshold1=50, threshold2=150)

# Morfológiai műveletek az élkiemelés finomításához
kernel_size = 3
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))
dilated_edges = cv2.dilate(edges, kernel, iterations=1)

height, width = gray_image.shape

recognized_data = pytesseract.image_to_boxes(dilated_edges, config='--psm 6', output_type=pytesseract.Output.DICT)

for i in range(len(recognized_data['char'])):
    char = recognized_data['char'][i]
    left = recognized_data['left'][i]
    bottom = height - recognized_data['top'][i]
    right = recognized_data['right'][i]
    top = height - recognized_data['bottom'][i]

    cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)

plt.figure(figsize=(25, 15))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Felismert karakterek')
plt.show()

recognized_text = ' '.join(recognized_data['char'])
print(recognized_text)
```

# Gradiens szűrés

```python
image = cv2.imread('Data/Handwriting/1.jpg', cv2.IMREAD_GRAYSCALE)

# Élszűrés gradiens alkalmazása
gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Élszűrés gradiens számítása
gradient = cv2.magnitude(gradient_x, gradient_y)

plt.figure(figsize=(10,10))
plt.imshow(gradient, cmap='gray')
plt.title('Élszűrés gradiens')
plt.axis('off')
plt.show()
```

```python
temp_image = 'temp_image.png'
cv2.imwrite(temp_image, gradient)

recognized_text = pytesseract.image_to_string(temp_image)

recognized_data = pytesseract.image_to_boxes(temp_image, config='--psm 3', output_type=pytesseract.Output.DICT)
highlighted_image = image.copy()

height = image.shape[0] 

for i in range(len(recognized_data['char'])):
    char = recognized_data['char'][i]
    left = recognized_data['left'][i]
    bottom = height - recognized_data['top'][i]
    right = recognized_data['right'][i]
    top = height - recognized_data['bottom'][i]

    cv2.rectangle(highlighted_image, (left, top), (right, bottom), (0, 255, 0), 2)

print(recognized_text)
plt.figure(figsize=(10, 10))
plt.imshow(highlighted_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()

os.remove(temp_image)
```

# AI Model

```python
train_data = pd.read_csv('Dataset/emnist-letters-train.csv')
test_data = pd.read_csv('Dataset/emnist-letters-test.csv')

# Címkék kinyerése és adatok előkészítése
y_train = train_data.iloc[:, 0].values.astype(int)
y_test = test_data.iloc[:, 0].values.astype(int)

x_train = train_data.iloc[:, 1:].values / 255.0
x_test = test_data.iloc[:, 1:].values / 255.0

# Modell létrehozása és tanítása
model = MLPClassifier(hidden_layer_sizes=(64, 64), max_iter=100, activation='relu', solver='adam', random_state=42)
model.fit(x_train, y_train)

# Tesztadatokon való értékelés
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print('Test accuracy:', accuracy)
```

```python
image = cv2.imread('Data/Handwriting/1.jpg', cv2.IMREAD_GRAYSCALE)
# Tesseract felismert szövege
recognized_text = pytesseract.image_to_string(image)

# Tesseract felismert karakterek pozícióinak kinyerése
recognized_data = pytesseract.image_to_boxes(image, config='--psm 6', output_type=pytesseract.Output.DICT)

# Kivágott karakterek tárolására használt tömb
cropped_characters = []

# Karakterek kivágása és tárolása
for i in range(len(recognized_data['char'])):
    char = recognized_data['char'][i]
    left = recognized_data['left'][i]
    bottom = image.shape[0] - recognized_data['top'][i]
    right = recognized_data['right'][i]
    top = image.shape[0] - recognized_data['bottom'][i]
    
    character_image = image[top:bottom, left:right]
    cropped_characters.append(character_image)

```

```python
prepared_images = []

for character_image in cropped_characters:
    resized_image = cv2.resize(character_image, (28, 28))
    normalized_image = resized_image / 255.0
    prepared_images.append(normalized_image.reshape(-1))

predictions = model.predict(prepared_images)

# Eredmények kiíratása
for i, prediction in enumerate(predictions):
    print(f"Karakter {i+1}: {prediction}")
```
